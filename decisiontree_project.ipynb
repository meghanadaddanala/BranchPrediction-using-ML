```python
# Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python

# load dataset
data = pd.read_csv('dataset60.csv')
data.head()
```

```python
data['college'].unique()  
```

```python
data.drop('college', axis=1, inplace=True)      
```

```python
data  
```

```python
data['caste'].unique() 
```

```python
data['caste'].replace(['oc', 'bc-a', 'bc-b', 'bc-c', 'bc-d', 'bc-e', 'sc', 'st'], [0,1,2,3,4,5,6,7], inplace=True)  
```

```python
data
```

```python
def binary_map(x):
  if x == 'm':
    return 0
  else:
    return 1


data['gender'] = data['gender'].apply(binary_map)
```

```python
data
```

```python
#Assign numeric values to branch column now. You can do dummy coding here as well
data['branch'].unique()
```

```python
data['branch'].replace(['cse', 'ece', 'eee', 'ce', 'me'], [0,1,2,3,4], inplace=True)
```

```python
data
```

```python
x = data.drop('branch', axis=1)
y = data['branch']
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.1, random_state=72)
```

```python
model = DecisionTreeClassifier()
model.fit(xtrain,ytrain)
print('Training Accuracy :', model.score(xtrain, ytrain))
print('Testing Accuracy :', model.score(xtest, ytest) )
```

```python
model.get_depth()
```

```python
model = DecisionTreeClassifier( criterion='entropy',max_depth=9)
model.fit(xtrain, ytrain)
print('Training Accuracy :', model.score(xtrain, ytrain))
print('Testing Accuracy :', model.score(xtest, ytest) )
```

```python
model = DecisionTreeClassifier( criterion='gini',max_depth=9)
model.fit(xtrain, ytrain)
print('Training Accuracy :', model.score(xtrain, ytrain))
print('Testing Accuracy :', model.score(xtest, ytest) )
```

```python
predicted = model.predict(xtrain)
xtrain['Actual output'] = ytrain
xtrain['Predicted output'] = predicted
xtrain
```

```python
import pickle
# Save the model
filename = 'model.pkl'
pickle.dump(model, open(filename, 'wb'))
```
